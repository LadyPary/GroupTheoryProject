**Definition 1** . $\displaystyle (\mathbb{Z} /n\mathbb{Z})$ Let $\displaystyle n$ be an integer greater than $\displaystyle 1$. The set$\displaystyle \ \{0,\ 1,\ .\ .\ .\ ,n\ −\ 1\}$ 
with the operation of$\displaystyle \ addition\ mod\ n$ will be called the set of integers $\displaystyle mod\ n\ $(or integers $\displaystyle modulo\ n$) and is denoted by $\displaystyle \mathbb{Z} /n\mathbb{Z}$. $\displaystyle (\mathbb{Z} /n\mathbb{Z} ,\ +)$ where $\displaystyle +$ is addition mod n forms a group.

**Definition 2** . $\displaystyle ( DihedralGroup( n))$ Let $\displaystyle P_{n}$  be the regular polygon in a plane with $\displaystyle n$ sides. The symmetries of $\displaystyle P_{n}$ are denoted by $\displaystyle D_{2n}$ which is called the $\displaystyle dihedral\ group$ of order$\displaystyle \ 2n$.
The elements of $\displaystyle D_{2n}$ are :

(a) rotations by $\displaystyle 0,\ 2π/n,\ .\ .\ .\ ,\ 2( n\ −\ 1) π/n\ radians$, and

(b) reflections: if $\displaystyle n$ is even, axes go through opposite vertices or opposite sides; if $\displaystyle n\ $is odd, axes go from a vertex to the midpoint of the opposite side.
If we let $\displaystyle r$ denote a rotation by $\displaystyle 2π/n$ $\displaystyle radians$ and we let $\displaystyle s$ denote any one of the reflections, then

$\displaystyle D_{2n} =\ \left\{e,\ r^{1} \ ,\ \ r^{2} \ ,\ .\ .\ .\ ,\ r^{\ ( n−1)} \ ,\ s\ ,r^{1} s,\ r^{2} s,\ .\ .\ .\ ,\ r^{( n−1)} s\ |\ r^{n} \ =\ s^{2} \ =\ e,\ sr^{1} \ =\ r^{( \ n−1) \ } s\right\} .$

*please note that in this project the terms $\displaystyle r^{i} \ $are displayed as $\displaystyle ri$.

**Definition 3** .$\displaystyle \ ( Sn) \ $ Let $\displaystyle n$ be a positive integer, and let$\displaystyle \ [ n] \ =\ \{1,\ 2,\ .\ .\ .\ ,\ n\}$. The collection of $\displaystyle 1-1$ and $\displaystyle onto$ functions from $\displaystyle [ n]$ to $\displaystyle [ n]$, that is, $\displaystyle Perm([ n])$, is called the $\displaystyle symmetric\ group\ of\ degree\ n$, and is denoted by $\displaystyle Sn$ . Hence 

$\displaystyle Sn\ =\ Perm([ n]) \ =\ \left\{f\ :\ [ n] \ \rightarrow \ [ n] \ |\ f\ is\ 1-1\ and\ onto\right\} .$
Elements of $\displaystyle Sn$ are called $\displaystyle permutations\ of\ [ n] .$

**Definition 4** .$\displaystyle ( The\ general\ linear\ group$). Let $\displaystyle n$ be a positive integer, and let $\displaystyle F$ be $\displaystyle \mathbb{Z} /p\mathbb{Z}$ for $\displaystyle p\ a\ prime$. Then the$\displaystyle general\ linear\ group\ $$\displaystyle \ GL( n,\ \mathbb{Z} /p\mathbb{Z} \ )$ is defined by

$\displaystyle GL( n,p) \ =\ \{n\ \times \ n\ invertible\ matrices\ with\ entries\ in\ \mathbb{Z} /p\mathbb{Z}\} .$

**Definition 5** .$\displaystyle ( The\ special\ linear\ group$). Let $\displaystyle n$ be a positive integer, and let $\displaystyle F$ be $\displaystyle \mathbb{Z} /p\mathbb{Z}$ for $\displaystyle p\ a\ prime$. Let $\displaystyle GL( n,\ p)$ be the general linear group of invertible$\displaystyle \ n\times n$ matrices with entries in $\displaystyle \mathbb{Z} /p\mathbb{Z}$ . Define

$\displaystyle SL( n,\ p) \ =\ \{A\ \in \ GL( n,\ p\ ) \ |\ det\ A\ =\ 1\} .$

Then $\displaystyle SL( n,\ p\ )$ is called the $\displaystyle special\ linear\ group$. 

**Definition 6** . $\displaystyle ( Set\ of\ Integers,addition\ modulo\ n)$ Let $\displaystyle n$ be an integer greater than $\displaystyle 1$. The set of arbitrary integers with the operation of $\displaystyle addition\ mod\ n$ can form a group if it satisfies the group axioms.

```
#@title **All the Functions** { display-mode: "form" ,run: "auto" }
#@markdown double click to see the python code.

import ipywidgets as widgets
from IPython.display import display
import numpy as np 
import itertools
import math


#printing functions
def print_one_matrix(a):
    if type(a)==tuple and type(a[0])==tuple:
        return np.array(a)
    return a

def func_dic_ele_for_form(groupele):
  i=1
  dic=dict()
  for each in groupele:
    dic[i]=each
    i+=1
  return dic

def func_print_ele_for_form(dict):
    for key in dict:
      print(key)
      print("--")
      print(print_one_matrix(dict[key]))
      print("=========")

def print_set_of_matrice(a):
    for each in a:
      if not type(each)==tuple:
        return a

      if not type(each[0])==tuple:
        return a

      else:  
        print("----")
        print(np.array(each))
    return "----"
    


def elements(GroupType,n,p=None):

    if GroupType=="Z/nZ":
        elementsofgroup=set(i for i in range(n))
        return elementsofgroup


    if GroupType=="DihedralGroup(n)":
        n =int(n/2)
        A =set("r"+ str(i) for i in range(1,n))
        B =set("r"+str(i)+"s" for i in range(1,n))

        elementsofgroup={"e"}|A|B|{"s"}

        return elementsofgroup

    if GroupType=="SymmetricGroup(n)":
        elementsofgroup=set(itertools.permutations(range(1, n+1)))
        return elementsofgroup

    if GroupType=="GL(n,p)":
        elementsofgroup=set()
        x =[y for y in range(p)]
        a=[j for j in itertools.product(x, repeat=n)]
        b={k for k in itertools.product(a, repeat=n)}

        for each in b:

            if not np.isclose((np.linalg.det(each)+1)%p,1):
                elementsofgroup.add(each)

        return elementsofgroup


    if GroupType=="SL(n,p)":
        elementsofgroup=set()
        x =[y for y in range(p)]
        a=[j for j in itertools.product(x, repeat=n)]
        b={k for k in itertools.product(a, repeat=n)}
        if p==2:
            for each in b:
                if np.isclose((np.linalg.det(each))%p,1):
                    elementsofgroup.add(each)
        else:
            for each in b:
                if np.isclose((np.linalg.det(each)+1)%p,2):
                    elementsofgroup.add(each)

        return elementsofgroup


def order(GroupType,n,p=None):

    if GroupType=="Z/nZ":
        return n

    if GroupType=="DihedralGroup(n)":
        return n


    if GroupType=="SymmetricGroup(n)":
        order=math.factorial(n)
        return order

    if GroupType=="GL(n,p)":
        order=1
        for i in range(0,n):
            order*=(p**n-p**i)
        return order

    if GroupType=="SL(n,p)":
        order=1
        for i in range(0,n):
            order*=(p**n-p**i)
        return int(order/(p-1))


def identity(GroupType,n,p=None):

    if GroupType=="Z/nZ":
        identityEle=0
        return identityEle

    if GroupType=="DihedralGroup(n)":
        identityEle="e"
        return identityEle


    if GroupType=="SymmetricGroup(n)":
        identityEle=tuple(i for i in range(1,n+1))
        return identityEle

    if GroupType=="GL(n,p)":
        identityEle=tuple(map(tuple,np.identity(n)))
        return identityEle

    if GroupType=="SL(n,p)":
        identityEle=tuple(map(tuple,np.identity(n)))
        return identityEle

    if GroupType=='Group(Set of Integers,addition mod n)':
        if 0 in Set:
            identityEle=0
        else:
            for ele in Set:
                if ele%n== 0:
                    identityEle=ele
                    break
        return identityEle


def composition(GroupType,n,p,ele1,ele2):

    if GroupType=="Z/nZ":
        comp=(ele1+ele2)%n
        return comp

    if GroupType=="DihedralGroup(n)":
        def left_s(b):
            if b=="s":
                temp="e"
            elif len(b)==2:
                temp= "r"+str(int(n/2)-int(b[-1]))+"s"
            else:
                temp= "r"+str(int(n/2)-int(b[1]))

            if temp[1]=="0":
                temp=temp[-1]

            if temp=="ss":
                temp="e"

            return temp

        def right_s(b):
            if b=="s":
                temp="e"
            elif len(b)==2:
                temp= b+"s"
            else:
                temp = b[:2]

            if temp=="ss":
                temp="e"

            return temp

        def both_rotations(a,b):
            temp= "r{}".format(int(((int(a[-1])+int(b[-1])))%(n/2)))

            if temp[1]=="0":
                temp="e"

            return temp


        def comp_s(a,b):

            temp=list(a+b+6*"0")
            i=0
            while len(temp)>6:

                if temp[i]=="0":
                    break

                if str("".join(temp[i:i+2]))=="sr":
                    temp[i:i+3]=left_s("".join(temp[i+1:i+3]))
                    i=0

                if str("".join(temp[i:i+2]))=="ss":
                    temp[i:i+3]="e"
                    i=0
                if i==1 and str("".join(temp[i-1:i+1]))=="ss":
                    temp[i-1:i+1]="e"
                    i=0

                if str(temp[i])+str(temp[i+2])=="rs":
                    temp[i:i+3]=right_s("".join(temp[i:i+2]))
                    i=0


                if str(temp[i])+str(temp[i+2])=="rr":
                    temp[i:i+4]=both_rotations("".join(temp[i:i+2]),"".join(temp[i+2:i+4]))
                    i=0

                if str(temp[i+1])=="e":
                    temp[i:i+2]=temp[i]
                    i=0

                if i==1 and str("".join(temp[i-1]))=="e":
                    temp[i-1:i+1]=temp[i]
                    i=0
                i+=1

            temp="".join([temp[i] for i in range(3) if temp[i] !="0" ])
            return temp


        #multiplying by identity 
        if ele1=="e":
            comp=ele2
        elif ele2=="e":
            comp=ele1

        #both are rotations
        elif len(ele1)==2 and len(ele2)==2:
            comp=both_rotations(ele1,ele2)

        ##multiplying by s 
        elif ele2=="s" and ele1=="s":
            comp="e"

        elif ele1=="s" :
            comp=left_s(ele2)

        elif ele2=="s":
            comp=right_s(ele1)


        else:
            comp=comp_s(ele1,ele2)

        return comp


    if GroupType=="SymmetricGroup(n)":

        comp=[0 for noom in range(n)]

        for ind in range(n):
            comp[ind]=ele1[ele2[ind]-1]

        comp=tuple(comp)

        return comp

    if GroupType=="GL(n,p)":
        temp=(np.matmul(ele1,ele2))%p
        comp=tuple(map(tuple, temp))

        return comp

    if GroupType=="SL(n,p)":
        temp=(np.matmul(ele1,ele2))%p
        comp=tuple(map(tuple, temp))

        return comp
        
    if GroupType=='Group(Set of Integers,addition mod n)':
        comp=(ele1+ele2)%n
        return comp



def inverse(GroupType,n,p,ele,setofelements):

    if GroupType=="Z/nZ":
        inv= (-ele)%n
        return inv

    if GroupType=="DihedralGroup(n)":
        for each in setofelements:
            if composition(GroupType,n,p,ele,each)=="e":
                temp=each
                break
        return temp

    if GroupType=="SymmetricGroup(n)":
        inv=[0 for l in range(n)]

        for ind in range(n):
            inv[ele[ind]-1]=ind+1

        inv=tuple(inv)
        return inv

    if GroupType=="GL(n,p)":
        for each in setofelements:
            if np.allclose(composition(GroupType,n,p,each,ele),np.identity(n)):
                return each 

    if GroupType=="SL(n,p)":
        for each in setofelements:
            if np.allclose(composition(GroupType,n,p,each,ele),np.identity(n)):
                return each 

    if GroupType=='Group(Set of Integers,addition mod n)':
      for i in setofelements:
          if (i+ele)%n==0:
              return i


def divisors(n):
    factors = {}
    nn = n
    i = 2
    while i*i <= nn:
        while nn % i == 0:
            factors[i] = factors.get(i, 0) + 1
            nn //= i
        i += 1
    if nn > 1:
        factors[nn] = factors.get(nn, 0) + 1

    primes = list(factors.keys())

    def generate(k):
        if k == len(primes):
            yield 1
        else:
            rest = generate(k+1)
            prime = primes[k]
            for factor in rest:
                prime_to_i = 1
                for _ in range(factors[prime] + 1):
                    yield factor * prime_to_i
                    prime_to_i *= prime

    yield from generate(0)


def power_for_Sn_and_Dn(GroupType,n,p,ele,powe):
    tempr=ele
    for times in range(powe-1):
        tempr=composition(GroupType,n,p,ele,tempr)
    return tempr

def element_order(GroupType,divisorsGO,n,p,ele,elementsofgroup,identity):
    #uses the fact that order of each element should divide the order of the group
    if GroupType=="Z/nZ":
        for each in divisorsGO:
            if ((each*ele)+1)%n==1:
                order=each
                break
        return order

    if GroupType=="DihedralGroup(n)":
        for eachp in divisorsGO:
            if power_for_Sn_and_Dn(GroupType,n,p,ele,eachp)==identity:
                order=eachp
                break

        return order

    if GroupType=="SymmetricGroup(n)": 
        for eachp in divisorsGO:
            if power_for_Sn_and_Dn(GroupType,n,p,ele,eachp)==identity:
                order=eachp
                break

        return order

    if GroupType=="GL(n,p)":
        for each in divisorsGO:
            power_matrice=np.linalg.matrix_power(ele,each)

            if np.allclose(power_matrice%p,np.identity(n)):
                order=each
                break

        return order


    if GroupType=="SL(n,p)":
        for each in divisorsGO:
            power_matrice=np.linalg.matrix_power(ele,each)

            if np.allclose(power_matrice%p,np.identity(n)):
                order=each
                break

        return order

    if GroupType=='Group(Set of Integers,addition mod n)':
        for each in divisorsGO:
            if ((each*ele)+1)%n==1:
                order=each
                break
        return order



def IsAbelain(GroupType,n):
#uses facts
    if GroupType=="Z/nZ":
        return True

    if GroupType=="DihedralGroup(n)":
        if n/2==0 or n/2==1:
            return True
        else:
            return False


    if GroupType=="SymmetricGroup(n)":
        if n==1 or n==2:
            return True
        else:
            return False

    if GroupType=="GL(n,p)":
        if n==0 or n==1:
            return True
        else:
            return False

    if GroupType=="SL(n,p)":
        if n==0 or n==1:
            return True
        else:
            return False

    if GroupType=='Group(Set of Integers,addition mod n)':
        return True


def centre_of_Group(GroupType,n,p,elementsofgroup):
#uses facts
    if GroupType=="Z/nZ":
        #it's abelian
        return elementsofgroup

    if GroupType=="DihedralGroup(n)":
    #uses the fact that if n/2 is odd ==> centre is trivial , if n/2 is even ==> centre is trivial+ a^(n/4)
        if (n/2)%2==0:
            temp= "r{}".format(int(n/4))
            centre={temp,"e"}
        else:
            centre={"e"}

        return centre


    if GroupType=="SymmetricGroup(n)":
    # n>2 ==> Then the center of Sn is trivial.
        if n==1 or n==2:
            centre= elementsofgroup
        else:
            centre ={tuple(i for i in range(1,n+1))}

        return centre


    if GroupType=="GL(n,p)":
    #The center of GL(n, F ). Let F be a field . Z(GL(n, F )), the center of the general
    #linear group, is {λI n | λ ∈ F }, the set of matrices that are constant
    #multiples of the identity matrix.
        centre=set()
        identity=np.identity(n)
        for num in range(1,p):
            temp=tuple(map(tuple,identity*num))
            centre.add(temp)

        return centre

    if GroupType=="SL(n,p)":
    #The center of GL(n, F ). Let F be a field . Z(SL(n, F )), the center of the general
    #linear group, is {λI n | λ ∈ F }, the set of matrices that are constant
    #multiples of the identity matrix.
        centre=set()
        identity=np.identity(n)
        for num in range(1,p):
            temp=tuple(map(tuple,identity*num))
            centre.add(temp)
        return centre

    if GroupType=='Group(Set of Integers,addition mod n)':
        return elementsofgroup

def IsGroup_GroupInt(Set,n):
    identity=None

    if 0 in Set:
        identity=0
    else:
        for ele in Set:
            if ele%n== 0:
                identity=ele
    if identity==None:
        return "Doesn't have an identity"
    else:
        GroupInfo["identity"]=identity

    
    for x in Set:
        for y in Set:
            if (x+y)%n not in Set and (x+y)%n+n!=n:
                return "{}+{} Doesn't have closure".format(x,y)

    for i in Set:
        if (-i)%n not in Set and i%n+n!=n:
            return "{} Doesn't have an inverse".format(i)

    return True
```

###**2.1. Group Order**
---
**Definition 2.1.1** (Order of a group). Let $\displaystyle G$ be a group.$\displaystyle \ G$ is a$\displaystyle \ finite\ group$ if it
has a finite number of elements; otherwise, $\displaystyle G$ is an$\displaystyle \ infinite\ group$. The number of
elements of $\displaystyle G$, denoted by $\displaystyle |G|$, is called the $\displaystyle order\ of\ G$.

It's clear that $\displaystyle |G|$ for $\displaystyle DihedralGroup( n)$ and ${\displaystyle (\mathbb{Z} /n\mathbb{Z} )}$ is $\displaystyle n$.

**Theorem 2.1.1** Let $\displaystyle n$ be a positive integer, Then the Order of $\displaystyle S( n)$ is $\displaystyle n!$ .

**proof 2.1.1:**
	${\displaystyle Sn\ =\ Perm([n])\ =\ \left\{f\ :\ [n]\ \rightarrow \ [n]\ |\ f\ is\ 1-1\ and\ onto\right\}}$ 
	and there are $\displaystyle n!$ bijections from ${\displaystyle \ [n]\ }$ to ${\displaystyle \ [n]\ }$. so ${\displaystyle |Sn|}$ has $\displaystyle n!$ elements therefore ${\displaystyle |Sn|=\ n!}$. $\displaystyle \square $



**Theorem 2.1.2**  Let $\displaystyle n$ be a positive integer, and let $\displaystyle p$ be a prime. Then
$\displaystyle |GL( n,\ p) |\ =\ \left( p^{n} \ −\ 1\right)\left( p^{n} \ −\ p\right)\left( p^{n} \ −\ p^{2} \ \right) \ \cdotp \ \cdotp \ \cdotp \ \left( p^{n} \ −\ p^{n−1} \ \right)$

$\displaystyle |SL( n,\ p) |\ =\frac{1}{p\ −\ 1} |GL( n,\ p) |\ .$

**proof 2.1.2:**
To count the number of invertible $\displaystyle n\times n$ matrices with entries in $\mathbb{Z} /p\mathbb{Z}$, we proceed one row at a time. There are $\displaystyle p^{n}$ choices for the first row, and we have to exclude the row of all zeros. Hence, there are $\displaystyle p^{n} \ −\ 1$ choices for the first row. For the second row, we have to exclude all the multiples of the first row. There are $\displaystyle p$ multiples of the first row—since
there are exactly $\displaystyle p\ $scalars—and hence the number of possibilities for the second row is $\displaystyle p^{n} \ −\ p$. Now for the third row, we have to exclude all the linear combinations of the first two rows. To find a linear combination of the first two rows, we have to choose two scalars, one to multiply the first row by and one to multiply by the second row. Hence there are $\displaystyle p^{2}$ possible linear combinations of the first two rows. So the number of rows that we can use for the third row is$\displaystyle \ p^{n} −p^{2}$ . Continuing in this way, we get:

$\displaystyle |GL( n,\ p) |\ =\ \left( p^{n} \ −\ 1\right)\left( p^{n} \ −\ p\right)\left( p^{n} \ −\ p^{2} \ \right) \ \cdotp \ \cdotp \ \cdotp \ \left( p^{n} \ −\ p^{n−1} \ \right)$

If a matrix is in $\displaystyle GL( n,\ p)$, then it is invertible and so its determinant cannot be zero. Hence, the choices for its determinant are $\displaystyle 1,\ 2,\ .\ .\ .,\ p\ −\ 1$. The point is that the matrices in$\displaystyle \ GL( n,\ p)$ are distributed evenly among these different possible determinants, and so the number of matrices in GL(n, p) that have determinant equal to $\displaystyle 1$ is exactly

$\displaystyle |SL( n,\ p) |\ =\frac{|GL( n,\ p) |}{p\ −\ 1} \ =\frac{\ \left( p^{n} \ −\ 1\right)\left( p^{n} \ −\ p\right)\left( p^{n} \ −\ p^{2} \ \right) \ \cdotp \ \cdotp \ \cdotp \ \left( p^{n} \ −\ p^{n−1} \ \right)}{p\ −\ 1}$.

Let $\displaystyle α$ and $\displaystyle β$ be two integers between $\displaystyle 1$ and $\displaystyle p\ −\ 1$. The only thing that remains to be shown is that the number of matrices with determinant $\displaystyle α$ is equal to the number of matrices with determinant$\displaystyle \ β$. Let $\displaystyle S_{1} \ =\ \{M\ \in \ GL( n,\ p) \ |\ det( M\ ) \ =\ α\} \ $and $\displaystyle S_{2} \ =\ \{M\ \in \ GL( n,\ p) \ |\ det( M\ ) \ =\ β\}$. Let $\displaystyle A\ $be the $\displaystyle n\ \times \ n\ $matrix defined by $\displaystyle A\ =\begin{bmatrix}
\beta /\alpha  & 0\\
0 & I_{n-1}
\end{bmatrix}$, where $\displaystyle I_{n-1}$ denotes the $\displaystyle ( n\ −\ 1) \ \times \ ( n\ −\ 1)$ identity matrix. Note that $\displaystyle det( A) \ =\ β/α$, and $\displaystyle A$ is invertible. Define$\displaystyle f:S\ _{1}\rightarrow \ S_{2} \ by\ f\ ( N\ ) \ =\ AN$.
	First note that if$\displaystyle \ N\ \in \ S_{1} \ $, then$\displaystyle \ det( AN\ ) \ =\ det( A) \ det( N\ ) \ =\frac{\beta \ }{\alpha } α=\beta $ and so $\displaystyle AN$ is indeed an element of $\displaystyle S_{2}$ Hence, $\displaystyle f$ is an actual function. If $\displaystyle AN_{1} \ =\ AN_{2}$ , then multiplying by $\displaystyle A^{−1}$ on the left, we get $\displaystyle N_{1} \ =\ N_{2}$ . Thus $\displaystyle f$ is $\displaystyle 1-1$. If $\displaystyle M$ is an arbitrary element of  $\displaystyle S_{2}$, then $\displaystyle det\left( A^{−1} M\ \right) \ =\ det\left( A^{−1\ }\right) \ det( M\ ) =\frac{\alpha \ }{\beta } \beta =\alpha $ and so  $\displaystyle A^{−1} M\in \ S_{1}$ . In addition, $\displaystyle f\ \left( A^{−1} M\ \right) \ =\ M$ and so $\displaystyle f$ is $\displaystyle onto$. $\displaystyle f$ is a$\displaystyle \ 1-1$, $\displaystyle onto$ map and so$\displaystyle \ |S_{1} \ |\ =\ |S_{2} \ |$, and the proof is complete.$\displaystyle \square $

###**2.3. The Group Centre**

**Theorem 2.4.1** Let $\displaystyle S_{n}$ be the symmetric group of order $\displaystyle n$ where $\displaystyle n\geq 3$. Then the centre of $\displaystyle \ S_{n}$ is trivial.

**Proof 2.4.1:**

So no non-identity elements of $\displaystyle S_{n}$ commute with all elements of $\displaystyle S_{n}$. Hence: $\displaystyle Z( S_{n}) =\{e\}$ $\displaystyle \square $

So, let$\displaystyle \ n\geq 3$. ###**2.5. Order of an Element**
---

**Definition 2.5.1** (Order of an element). Let $\displaystyle G$ be a group, and let $\displaystyle a\ \in \ G.$ The $\displaystyle order\ of\ a$, denoted by $\displaystyle o( a)$, is defined to be$\displaystyle \ |\langle a\rangle |.$ Equivalently, $\displaystyle o( a) \ $is the smallest positive integer $\displaystyle d$ such that $\displaystyle a^{d} \ =\ e.$ 


**Theorem 2.5.1** (Lagrange). Let $\displaystyle G\ $be a finite group. Let $\displaystyle H\ \leq \ G$. Then $\displaystyle |G|\ =\ |H|\ \cdotp \ |G\ :\ H|.$In particular,$\displaystyle \ |H|\ divides\ |G|$.

**Proof 2.5.2:** Let $\displaystyle H$ act on $\displaystyle G$ by translation. The orbits are the right cosets, each of these orbits have size$\displaystyle \ |H|$, and the orbits partition $\displaystyle G$. The number of orbits is$\displaystyle \ |G\ :\ H|$ and hence $\displaystyle G$ is partitioned into$\displaystyle \ |G\ :\ H|\ $sets, each having$\displaystyle \ |H|\ $elements. $\displaystyle \square $



**Corollary 2.5.3.** Let$\displaystyle \ |G|\ < \ \infty $, a group, and let $\displaystyle a\ \in \ G$. Then $\displaystyle o( a)$ divides$\displaystyle \ |G|$.

**Proof 2.5.3.** The cyclic group generated by$\displaystyle \ a$, that is $\displaystyle \langle a\rangle $, is a subgroup of G. Hence Lagrange’s theorem implies that$\displaystyle \ |\langle a\rangle |$ $\displaystyle divides\ |G|$. But, $\displaystyle o(a) \ =\ \ |\langle a\rangle |$, and the proof is complete.$\displaystyle \square $
